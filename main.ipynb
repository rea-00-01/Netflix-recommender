```python
movies=movies[['Id','Title','Genres','Overview']]
```

```python
movies['tags']=movies['Genres']+' '+movies['Overview']
movies.head()
```

```python
movies=movies.drop(columns=['Genres','Overview'])
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#to convert text into vector
cv=CountVectorizer(max_features=5000 ,stop_words='english')
cv.fit_transform(movies['tags'].astype('U')).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=5000,stop_words='english')
vector=tfidf.fit_transform(movies['tags'].values.astype('U')).toarray()
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim=cosine_similarity(vector)
sim
```

```python
#ab isko enumerate krdenge kyuki we want to find top 5 results ,enumerate isko sbke sath iterate krdeg
sorted(list (enumerate(sim[2])),reverse=True)
#here sorting is basd on index [0]
sorted(list(enumerate(sim[2])),reverse=True,key=lambda x:x[1])
#now it will sort by index[1] jo ki sim_score h
```

```python
distance=sorted(list(enumerate(sim[2])),reverse=True,key=lambda x:x[1])
for i in distance[0:5]:
    print(i)
#it also gives us the movie from which we are looking for recommendation we can remove it ...distance[1:6] and then we can print the movie name using index
```

```python
for i in distance[0:6]:
    print(movies.iloc[i[0]].Title)
```

```python
def recommend(movie_name):
    movie_index=movies[movies['Title']==movie_name].index[0]
    distance=sorted(list(enumerate(sim[movie_index])),reverse=True, key=lambda x:x[1])
    for i in distance[1:6]:
        print(movies.iloc[i[0]].Title)
```

```python
recommend('The Dark Knight Rises')
```
